## Hyper SQL (based on PostgreSQL)
In databases like PostgreSQL (which Hyper is based on), `public` is the default "drawer" where all tables are placed unless you specifically create a new one. When you export data from Python using `pantab`, it automatically puts your table into this `public` schema so Tableau can find it easily.

```sql
\attach "ecommerce_cleaned.hyper" my_data
```

```sql
SELECT "Category", SUM("Amount") AS "Total_Sales"
FROM "my_data"."public"."AmazonSales"
GROUP BY 1;    -- 1 refers back to "Category"
```

```sql
-- Identify "High Value" Categories (Ordering)
SELECT "Category", SUM("Amount") AS "Total_Sales"
FROM "my_data"."public"."AmazonSales"
GROUP BY 1
ORDER BY 2 DESC;   -- put the biggest numbers at the top
```

## <span style="color:purple">`COUNT(*)` is a function used to count the total number of rows (records) in a specific group or table </span>

```sql
-- Check for missing data (filtering
SELECT "Category", COUNT(*) AS "Missing_Rows"
FROM "my_data"."public"."AmazonSales"
WHERE "Amount" IS NULL -- This counts how many rows in each category are still missing an Amount
GROUP BY 1;
```

```sql
-- Calculate average order
SELECT "Category", COUNT(*) AS "Order_Count",
    AVG("Amount") AS "Average_Sale",
    MAX("Amount") AS "Highest_Single_Sale"
FROM "my_data"."public"."AmazonSales"
GROUP BY 1
ORDER BY 3 DESC;
```
